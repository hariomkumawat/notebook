```python
def my_split(string, separator=None, maxsplit=-1):
    """
    Split a string into a list of substrings, using the given separator.

    Args:
    string (str): The string to split.
    separator (str, optional): The separator to use for splitting the string. If not provided, whitespace is used as the separator.
    maxsplit (int, optional): The maximum number of splits to perform. If not provided, all possible splits are performed.

    Returns:
    list: A list of substrings.

    Examples:
    >>> my_split("Hello World")
    ['Hello', 'World']

    >>> my_split("Hello-World", '-')
    ['Hello', 'World']

    >>> my_split("Hello-World-How-Are-You", '-', 2)
    ['Hello', 'World', 'How-Are-You']
    """
    if separator is None:
        separator = ' '
    parts = []
    current_part = ''
    num_splits = 0
    for char in string:
        if num_splits == maxsplit:
            current_part += char
            continue
        if char == separator:
            parts.append(current_part)
            current_part = ''
            num_splits += 1
        else:
            current_part += char
    parts.append(current_part)
    return parts
print(my_split('hello world'))
print(my_split('hello-world','-'))
print(my_split("Hello-World-How-Are-You", '-', 2))
```
